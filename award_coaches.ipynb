```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_column',None)
pd.set_option('display.max_row',None)
```

```python
#loading the dataset
df=pd.read_csv("basketball_awards_coaches.csv")
df
```

```python
#data set info
df.info()
```

```python
#checking for duplicates
df_dup=df.duplicated()
df_dup_df=pd.DataFrame(df_dup,columns=['duplicate'])
df_dup_df
```

```python
#checking number of columns
df.columns
```

```python
#renaming column lgID for cosistency and clarity
award=df.rename(columns={'lgID':'leagueID'})
award
```

```python
#checking for missing values
award_null=award.isnull().sum()
award_null_award=pd.DataFrame(award_null,columns=['missing values'])
award_null_award
```

```python
#checking for percentage of missing value.
float_percentage = []
string_percentage = []
for key, values  in award_null.items():
    percent_missing=(values/60)*100
    string_percentage.append(str(percent_missing)+'%')
    float_percentage.append(percent_missing)
missing_null=pd.DataFrame(index=award_null.keys(),columns=['percentage null'])
missing_null['percentage null']=string_percentage
missing_null
```

```python
#confirming the data size
award.shape
```

```python
#droping nulls
award_drop=award.dropna()
award_drop
```

```python
#checking for nulls after droping nulls
award_drop.shape
```

inference
It is seen clearly that if the null in note column is to be droped the entire data set would be gone as more than 95% of the note column is a null also some of the null values in the note column are neccessarily not applicable.

(1) Hence instead of dropping the null values they will be replaced with not applicable.

```python
#replace nulls in note column with not applicable
award['note']=award['note'].fillna('not applicable')
award
```
