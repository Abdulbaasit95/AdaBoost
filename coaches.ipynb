```python
#checking for number of amnt of missing data.
amnt_of_missing_data=ren.isnull().sum()
amnt_of_missing_data_ren=pd.DataFrame(amnt_of_missing_data,columns=['amnt of missing data'])
amnt_of_missing_data_ren
```

```python
#percentage of missing values

string_percentage = []
float_percentage = []
for key, values in amnt_of_missing_data.items():
    percent_missing=(values/1689)*100
    float_percentage.append(percent_missing)
    string_percentage.append(str(percent_missing) +'%')
missing_per=pd.DataFrame(index= amnt_of_missing_data.keys(),columns=['percentage missing'])
missing_per['percentage missing']= string_percentage
missing_per
```
